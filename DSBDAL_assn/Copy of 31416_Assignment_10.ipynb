### 3. Create a boxplot for each feature in the dataset.

```python
plt.rcParams['figure.figsize'] = (8, 8)
sns.boxplot(x='sepal.length', data=df)
```

```python
sns.boxplot(x='sepal.width', data=df)
```

```python
sns.boxplot(x='petal.length', data=df)
```

```python
sns.boxplot(x='petal.width', data=df)
```

```python
from scipy import stats

q1_sw = df['sepal.width'].quantile(0.25)
q3_sw = df['sepal.width'].quantile(0.75)
iqr_sw = stats.iqr(df['sepal.width'])
lower = q1_sw - 1.5*iqr_sw
upper = q3_sw + 1.5*iqr_sw
df = df.loc[(df['sepal.width'] > lower) & (df['sepal.width'] < upper)]
```

```python
sns.boxplot(x='sepal.width', data=df)
```

```python
# Import label encoder
from sklearn import preprocessing

# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()

# Encode labels in column 'variety'.
df['variety'] = label_encoder.fit_transform(df['variety'])
df['variety'].unique()
```
